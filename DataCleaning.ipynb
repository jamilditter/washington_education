```python
import pandas as pd
import requests
```

Here, we're loading in all the datasets we're going to use from the State of Washington Open Data API, using Report Card data.

```python
url = 'https://data.wa.gov/resource/rxjk-6ieq.json?$limit=243167'

response = requests.get(url)
data = response.json()

school_demo_2014to2026 = pd.DataFrame(data)
```

```python
school_demo_total_filtered = school_demo_2014to2026[school_demo_2014to2026['gradelevel']=="All Grades"]
school_demo_total_filtered = school_demo_total_filtered[school_demo_total_filtered['schoolname']!="District Total"]
school_demo_total_filtered
```
