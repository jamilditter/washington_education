```python
school_test_total = pd.concat([school_test_2014to2021, school_test_2022, school_test_2023, school_test_2024])
```

```python
school_test_total = school_test_total[school_test_total['gradelevel'] == 'All Grades']
school_test_total = school_test_total[school_test_total['schoolname'] == 'District Total']

school_test_total
```

Finally, we merge the two datasets together with a leftjoin on school year and districtorganizationid.

```python
school_data = pd.merge(school_test_total, school_demo_total_filtered, how="left", on=['schoolyear', 'districtorganizationid'])
```

```python
school_data.columns
```

```python
school_data = school_data.rename(columns={
    'organizationlevel_x': 'organizationlevel',
    'county_x': 'county',
    'esdname_x': 'esdname',
    'esdorganizationid_x': 'esdorganizationid',
    'districtcode_x': 'districtcode',
    'districtname_x': 'districtname',
    'schoolname_x': 'schoolname',
    'gradelevel_x': 'gradelevel',
    'dat_x': 'dat',
    'dataasof_x': 'dataasof',
    'schoolcode_x': 'schoolcode',
    'currentschooltype_x': 'currentschooltype'
})
```

```python
columns_to_drop = [
    'organizationlevel_y', 'county_y', 'esdname_y', 'esdorganizationid_y',
    'districtcode_y', 'districtname_y', 'schoolname_y', 'gradelevel_y',
    'dataasof_y', 'dat_y', 'schoolcode_y', 'schoolorganizationid_y',
    'currentschooltype_y'
]

school_data = school_data.drop(columns=columns_to_drop)
```
